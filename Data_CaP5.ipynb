#### Задание 1.
Написать программу, которая собирает входящие письма из своего или тестового почтового ящика и сложить данные о письмах в базу данных (от кого, дата отправки, тема письма, текст письма полный).

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from pymongo import MongoClient
from pprint import pprint
import time
```

```python
chrome_options = Options()
chrome_options.add_argument('start-maximized')
```

```python
driver = webdriver.Chrome(options=chrome_options)
driver.get('https://mail.ru/')

login = driver.find_element_by_id('mailbox:login')
login.send_keys('study.ai_172@mail.ru')
login.send_keys(Keys.ENTER)

time.sleep(0.5)

password = driver.find_element_by_id('mailbox:password')
password.send_keys('NextPassword172')
password.send_keys(Keys.ENTER)

time.sleep(10)
```

```python
letter_links = []

for i in range(14):
    
    time.sleep(10) # Обработать все письма получилось только с таким значением, не ниже
    
    links = driver.find_elements_by_class_name('js-letter-list-item')
    
    for link in links:
        letter_link = link.get_attribute('href')
        if letter_link not in letter_links:
            letter_links.append(letter_link)
        else:
            continue
    
    letters = driver.find_elements_by_class_name('js-letter-list-item')
    actions = ActionChains(driver)
    actions.move_to_element(letters[-1])
    actions.perform()
```

```python
letters = []

for letter in letter_links:
    
    data = driver.get(letter)
    
    time.sleep(5)
    
    letters_data = {}
    
    try:
        sender = driver.find_element_by_class_name('letter-contact')
        sender = sender.get_attribute('title') # Получаем контакт отправителя
    except:
        pass

    try:
        date = driver.find_element_by_class_name('letter__date').text # Получаем дату и время отправки письма
    except:
        pass

    try:
        subject = driver.find_element_by_class_name('thread__subject-line').text # Получаем тему письма
    except:
        pass

    try:
        body = driver.find_element_by_class_name('letter__body').text # Получаем текст письма
    except:
        pass

    letters_data['sender'] = sender
    letters_data['date'] = date
    letters_data['subject'] = subject
    letters_data['body'] = body
    letters.append(letters_data)
```

```python
driver.quit()
```

```python
pprint(letters)
```

```python
client = MongoClient('localhost',27017)
db = client['letters_db']
letters_db = db.letters_db
```

```python
def add_letters_to_db(letters):
    
    letters_db.insert_many(letters)

    return True
```

```python
add_letters_to_db(letters)
```

#### Задание 2.
Написать программу, которая собирает «Хиты продаж» с сайта техники mvideo и складывает данные в БД. Магазины можно выбрать свои. Главный критерий выбора: динамически загружаемые товары.

```python
chrome_options = Options()
chrome_options.add_argument('start-maximized')
```

```python
driver = webdriver.Chrome(options=chrome_options)
driver.get('https://www.mvideo.ru/')

time.sleep(10)

# Поиск нужного блока
data = driver.find_elements_by_class_name('accessories-product-list')
data = data[2]
hits = data.find_elements_by_class_name('gallery-list-item')

hits_ = []

i = 0

while i < 4:
    
    for hit in hits:
        
        info = hit.find_element_by_tag_name('a')
        info = info.get_attribute('data-product-info').replace('\n', '').replace('\t', '') # Собираем информацию о товарах
        info = eval(info) # Переводим из строки в словарь
        hits_.append(info) # Добавляем в список для дальнейшего сбора товаров в базу данных

        # Обработка блока, состоящего из 4 товаров
        item = data.find_elements_by_tag_name('li')
        actions = ActionChains(driver)
        actions.move_to_element(item[3])
        actions.perform()
    
    # Нажимаем на кнопку
    button = driver.find_elements_by_class_name('accessories-carousel-wrapper')
    button = button[2]
    button = button.find_elements_by_tag_name('a')
    button = button[-1]
    button.click()

    i += 1
```

```python
driver.quit()
```

```python
pprint(hits_)
```

```python
client = MongoClient('localhost',27017)
db = client['mvideo_hits_db']
mvideo_hits = db.mvideo_hits
```

```python
def add_mvideo_hits_to_db(hits_):
    
    mvideo_hits.insert_many(hits_)

    return True
```
