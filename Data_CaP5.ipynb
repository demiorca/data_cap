```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from pymongo import MongoClient
from pprint import pprint
import time
```

```python
chrome_options = Options()
chrome_options.add_argument('start-maximized')
```

```python
driver = webdriver.Chrome(options=chrome_options)
driver.get('https://mail.ru/')

login = driver.find_element_by_id('mailbox:login')
login.send_keys('study.ai_172@mail.ru')
login.send_keys(Keys.ENTER)

time.sleep(0.5)

password = driver.find_element_by_id('mailbox:password')
password.send_keys('NextPassword172')
password.send_keys(Keys.ENTER)

time.sleep(10)
```

```python
letter_links = []

for i in range(14):
    
    time.sleep(10) # –û–±—Ä–∞–±–æ—Ç–∞—Ç—å –≤—Å–µ –ø–∏—Å—å–º–∞ –ø–æ–ª—É—á–∏–ª–æ—Å—å —Ç–æ–ª—å–∫–æ —Å —Ç–∞–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, –Ω–µ –Ω–∏–∂–µ
    
    links = driver.find_elements_by_class_name('js-letter-list-item')
    
    for link in links:
        letter_link = link.get_attribute('href')
        if letter_link not in letter_links:
            letter_links.append(letter_link)
        else:
            continue
    
    letters = driver.find_elements_by_class_name('js-letter-list-item')
    actions = ActionChains(driver)
    actions.move_to_element(letters[-1])
    actions.perform()
```

```python
letters = []

for letter in letter_links:
    
    data = driver.get(letter)
    
    time.sleep(5)
    
    letters_data = {}
    
    try:
        sender = driver.find_element_by_class_name('letter-contact')
        sender = sender.get_attribute('title') # –ü–æ–ª—É—á–∞–µ–º –∫–æ–Ω—Ç–∞–∫—Ç –æ—Ç–ø—Ä–∞–≤–∏—Ç–µ–ª—è
    except:
        pass

    try:
        date = driver.find_element_by_class_name('letter__date').text # –ü–æ–ª—É—á–∞–µ–º –¥–∞—Ç—É –∏ –≤—Ä–µ–º—è –æ—Ç–ø—Ä–∞–≤–∫–∏ –ø–∏—Å—å–º–∞
    except:
        pass

    try:
        subject = driver.find_element_by_class_name('thread__subject-line').text # –ü–æ–ª—É—á–∞–µ–º —Ç–µ–º—É –ø–∏—Å—å–º–∞
    except:
        pass

    try:
        body = driver.find_element_by_class_name('letter__body').text # –ü–æ–ª—É—á–∞–µ–º —Ç–µ–∫—Å—Ç –ø–∏—Å—å–º–∞
    except:
        pass

    letters_data['sender'] = sender
    letters_data['date'] = date
    letters_data['subject'] = subject
    letters_data['body'] = body
    letters.append(letters_data)
```

```python
driver.quit()
```

```python
pprint(letters)
```

```python
chrome_options = Options()
chrome_options.add_argument('start-maximized')
```

```python
driver = webdriver.Chrome(options=chrome_options)
driver.get('https://www.mvideo.ru/')

time.sleep(10)

# –ü–æ–∏—Å–∫ –Ω—É–∂–Ω–æ–≥–æ –±–ª–æ–∫–∞
data = driver.find_elements_by_class_name('accessories-product-list')
data = data[2]
hits = data.find_elements_by_class_name('gallery-list-item')

hits_ = []

i = 0

while i < 4:
    
    for hit in hits:
        
        info = hit.find_element_by_tag_name('a')
        info = info.get_attribute('data-product-info').replace('\n', '').replace('\t', '') # –°–æ–±–∏—Ä–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç–æ–≤–∞—Ä–∞—Ö
        info = eval(info) # –ü–µ—Ä–µ–≤–æ–¥–∏–º –∏–∑ —Å—Ç—Ä–æ–∫–∏ –≤ —Å–ª–æ–≤–∞—Ä—å
        hits_.append(info) # –î–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ —Å–±–æ—Ä–∞ —Ç–æ–≤–∞—Ä–æ–≤ –≤ –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö

        # –û–±—Ä–∞–±–æ—Ç–∫–∞ –±–ª–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–µ–≥–æ –∏–∑ 4 —Ç–æ–≤–∞—Ä–æ–≤
        item = data.find_elements_by_tag_name('li')
        actions = ActionChains(driver)
        actions.move_to_element(item[3])
        actions.perform()
    
    # –ù–∞–∂–∏–º–∞–µ–º –Ω–∞ –∫–Ω–æ–ø–∫—É
    button = driver.find_elements_by_class_name('accessories-carousel-wrapper')
    button = button[2]
    button = button.find_elements_by_tag_name('a')
    button = button[-1]
    button.click()

    i += 1
```
